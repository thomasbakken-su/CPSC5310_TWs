```python
%matplotlib notebook
# Python ≥3.5 is required
import sys
assert sys.version_info >= (3, 5)

# Scikit-Learn ≥0.20 is required
import sklearn
assert sklearn.__version__ >= "0.20"

# Common imports
import numpy as np
import pandas as pd
import seaborn as sn
import os

from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
from sklearn.model_selection import train_test_split
from sklearn.datasets import make_classification, make_blobs
from matplotlib.colors import ListedColormap
from sklearn.datasets import load_breast_cancer


# To plot pretty figures
%matplotlib inline
import matplotlib as mpl
import matplotlib.pyplot as plt
mpl.rc('axes', labelsize=14)
mpl.rc('xtick', labelsize=12)
mpl.rc('ytick', labelsize=12)


# to make this notebook's output stable across runs
np.random.seed(42)

# Where to save the figures
PROJECT_ROOT_DIR = "."

CHAPTER_ID = 'Naive Bayesian'
IMAGES_PATH = os.path.join(PROJECT_ROOT_DIR, "images", CHAPTER_ID)

os.makedirs(IMAGES_PATH, exist_ok=True)

def save_fig(fig_id, tight_layout=True, fig_extension="png", resolution=300):
    path = os.path.join(IMAGES_PATH, fig_id + "." + fig_extension)
    print("Saving figure", fig_id)
    if tight_layout:
        plt.tight_layout()
    plt.savefig(path, format=fig_extension, dpi=resolution)
```

### 1. Sythetic Datasets

```python
# synthetic dataset for classification (binary)

cmap_bold = ListedColormap(['#FFFF00', '#00FF00', '#0000FF','#000000'])

plt.figure()
plt.title('Sample binary classification problem with two informative features')

X, y = make_classification(n_samples = 100, n_features=2,
                                n_redundant=0, n_informative=2,
                                n_clusters_per_class=1, flip_y = 0.1,
                                class_sep = 0.5, random_state=0)

plt.scatter(X[:, 0], X[:, 1], marker= 'o', c=y, s=50, cmap=cmap_bold)
plt.show()
```

## Naive Bayes classifiers

```python
from sklearn.naive_bayes import GaussianNB
from adspy_shared_utilities import plot_class_regions_for_classifier

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)

nbclf = GaussianNB().fit(X_train, y_train)

plot_class_regions_for_classifier(nbclf, X_train, y_train, X_test, y_test,
                                 'Gaussian Naive Bayes classifier: Dataset 1')
```

```python
nbclf.score(X_test, y_test)
```

```python
y_pred = nbclf.predict(X_test)

# plot a confusion matrix
confusion_mat = confusion_matrix(y_test, y_pred)

print(confusion_mat)
```

```python
# Print classification report
target_names = ['Class 0', 'Class 1']

result_metrics = classification_report(y_test, y_pred, target_names=target_names)

print(result_metrics)
```

## ROC Curve

```python
from sklearn.metrics import roc_curve, auc
y_score = nbclf.predict_proba(X_test)

false_positive_rate, true_positive_rate, thresholds = roc_curve(y_test, y_score[:,1])

roc_auc = auc(false_positive_rate, true_positive_rate)
print('Accuracy = ', roc_auc)

count = 1

# Get different color each graph line
colorSet = ['navy', 'greenyellow', 'deepskyblue', 'darkviolet', 'crimson', 
            'darkslategray', 'indigo', 'brown', 'orange', 'palevioletred', 'mediumseagreen',
            'k', 'darkgoldenrod', 'g', 'midnightblue', 'c', 'y', 'r', 'b', 'm', 'lawngreen'
            'mediumturquoise', 'lime', 'teal', 'drive', 'sienna', 'sandybrown']
color = colorSet[count-1]

# Plotting
plt.title('ROC')
plt.plot(false_positive_rate, true_positive_rate, c=color, label=('Accuracy = %0.2f'%roc_auc))
plt.legend(loc='lower right', prop={'size':8})
plt.plot([0,1],[0,1], color='lightgrey', linestyle='--')
plt.xlim([-0.05,1.0])
plt.ylim([0.0,1.05])
plt.ylabel('True Positive Rate')
plt.xlabel('False Positive Rate')
plt.show()
```

```python
import pandas as pd

iris = pd.read_csv("./data/iris.csv")
target_names = list(iris.Name.unique())
y = iris.loc[:,'Class']
X = iris.loc[:, 'SepalLength':'PetalWidth']
X_norm = (X - X.min(axis=0)) / (X.max(axis=0) - X.min(axis=0))
X = pd.concat([X, X_norm], axis=1)

X_train, X_test, y_train, y_test = train_test_split(X, y)

nbclf = GaussianNB().fit(X_train.iloc[:,0:3], y_train)
print(nbclf.score(X_test.iloc[:,0:3], y_test))

y_pred = nbclf.predict(X_train.iloc[:,0:3])
confusion_mat = confusion_matrix(y_train, y_pred)
print(confusion_mat)

result_metrics = classification_report(y_train, y_pred, target_names=target_names)
print(result_metrics)
```

```python
#normalized data
nbclf_norm = GaussianNB().fit(X_train.iloc[:,4:], y_train)
print(nbclf_norm.score(X_test.iloc[:,4:], y_test))

y_pred_norm = nbclf_norm.predict(X_train.iloc[:,4:])
confusion_mat = confusion_matrix(y_train, y_pred_norm)
print(confusion_mat)

result_metrics = classification_report(y_train, y_pred_norm, target_names=target_names)
print(result_metrics)
```

```python
#applying k-fold
from sklearn.model_selection import KFold
nbclf_reg = GaussianNB()
nbclf_norm = GaussianNB()

kf = KFold(n_splits=10, random_state=None, shuffle=True)


count = 1
for train_index, test_index in kf.split(X):
    X_train_reg, X_test_reg = X.iloc[train_index, 0:4], X.iloc[test_index,0:4]
    X_train_norm, X_test_norm = X.iloc[train_index, 4:], X.iloc[test_index,4:]
    y_train, y_test = y[train_index], y[test_index]
    
    nbclf_reg.fit(X_train_reg, y_train)
    nbclf_norm.fit(X_train_norm, y_train)
    
    y_pred_reg = nbclf_reg.predict(X_train_reg)
    y_pred_norm = nbclf_norm.predict(X_train_norm)
    
    confusion_mat_reg = confusion_matrix(y_train, y_pred_reg)
    confusion_mat_norm = confusion_matrix(y_train, y_pred_norm)
    
    result_metrics_reg = classification_report(y_train, y_pred_reg, target_names=target_names)
    result_metrics_norm = classification_report(y_train, y_pred_norm, target_names=target_names)
    
    print("\n\nFold Number", count)
    count += 1
    
    print("\nRegular Score:", nbclf_reg.score(X_test_reg, y_test))
    print("\nRegular Confusion Matrix:\n", confusion_mat_reg)
    print("\nRegular Classification Report:\n", result_metrics_reg)
    print("\nNormalized Score:", nbclf_norm.score(X_test_norm, y_test))
    print("\nNormalized Confusion Matrix:\n", confusion_mat_norm)
    print("\nNormalized Classification Report:\n", result_metrics_norm)

```
